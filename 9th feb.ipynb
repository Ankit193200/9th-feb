# Q3: What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

Multiple inheritance in Python is when a class is derived from more than one base classes. Here's a simple example:

python
Copy code
class A:
    def method_a(self):
        print("Method A from class A")

class B:
    def method_b(self):
        print("Method B from class B")

class C(A, B):
    def method_c(self):
        print("Method C from class C")


obj_c = C()


obj_c.method_a()
obj_c.method_b()
obj_c.method_c()
# Q4: What are getter and setter in Python? Create a class with getter and setter methods.

Getter and setter methods are used to get and set the values of private instance variables, respectively.

python
Copy code
class MyClass:
    def __init__(self):
        self._my_variable = None

    
    def get_my_variable(self):
        return self._my_variable

    
    def set_my_variable(self, value):
        self._my_variable = value


obj = MyClass()
obj.set_my_variable(42)
print(obj.get_my_variable())  
Q5: What is method overriding in Python? Write a Python code to demonstrate method overriding.

Method overriding occurs when a subclass provides a specific implementation for a method that is already defined in its superclass.

python
Copy code
class Parent:
    def show(self):
        print("Parent class method")

class Child(Parent):
    def show(self):
        print("Child class method")


parent_obj = Parent()
child_obj = Child()


parent_obj.show()  
child_obj.show()   
In the example above, the show method in the Child class overrides the show method in the Parent class. When calling show on an instance of Child, it uses the overridden method in the child class.
